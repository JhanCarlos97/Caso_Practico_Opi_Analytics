## Online Retailing

Online retailing continues to grow at a rapid pace. A recent report by Forrester Research estimates that online sales in 2013 will be 262 billion dolars, a rise of 13% over 2012—reaching 8% of all US retail sales—and that online sales will continue to grow at a compound annual growth rate of 9% between 2012 and 2017. 1 While retailers such as Amazon operate exclusively online, legacy retail chains have had to build online channels alongside their
brick-and-mortar (B&M) operations. This has created a number of challenges as major retailers such as Home Depot, Target, Toys “R” Us, Crate and Barrel, Macy’s, and Nordstrom have struggled with the most effective way to integrate online and B&M channels.

## Home and Kitchen

Home and Kitchen, a chain with 84 stores in North America, 67 of which are in the United States and 17 in Canada, is one such retailer. The company has had a long and established presence through its B&M stores and introduced its online store in 2007. Online orders ship to anywhere in the United States or Canada from a central warehouse.
Since 2007, the company has had two main divisions: online and brick & mortar, each operated as a profit center. Kristine Schwarz was the head of the online channel and has been in that position since its inception. After significant growth over the first five years, the company’s online sales had started to stabilize. So like many retailers, Home and Kitchen was vigorously pursuing mechanisms to reinvigorate growth in online sales.

## The BOPS Initiative

Schwarz was always on the lookout for potential new initiatives to boost online sales. In 2010, after significant internal discussion and debate, Schwarz and her team launched a new strategic initiative called Buy Online Pickup at Store (BOPS). The BOPS program added a new delivery option for customers purchasing online. In addition to the usual shipping options, customers were now given the option of picking up their purchases at a nearby store within a given time window. In this way, customers could “book” the product online and then pick it up at a nearby store. This enabled customers to save on shipping fees and take their purchase home the same day. Schwarz felt BOPS would have great appeal.

Because there had always been some internal rivalry between the online and B&M channel heads, Schwarz had to closely coordinate with her B&M counterpart to ensure a smooth introduction of BOPS. To make matters worse, BOPS was operationally complex. It required that the B&M division provide the online division with real-time store availability data for all stores, and the B&M division had to develop new in-store processes to reliably reserve products in a specific store once a transaction had taken place online. The B&M head had resisted the BOPS program at first given the extent of effort it required for his division and the added burden it imposed on store personnel, who would need to service BOPS online orders in addition to their regular duties managing in-store sales. Adding insult to injury, BOPS sales would be booked under the online division’s P&L. Yet despite some grumbling in the B&M division, this was a major growth initiative at Home and Kitchen and everyone in the company had a stake in its success.

After a year of intense effort to integrate the information systems, BOPS was launched on October 11, 2011. The Canadian operation had been more reluctant to move ahead with BOPS and by the October launch date its information systems were still not ready to provide real-time inventory status. As a result, the BOPS initiative was only launched in the United States.

## Sales Results after BOPS

It was now April 22, 2012, and Schwarz was thinking ahead to the next executive meeting that would take place in two weeks. One of the main agenda items at the meeting was an evaluation of the BOPS initiative. The fate of the initiative was on the line and Schwarz was worried. Many sales associates in the brick & mortar channel had been complaining about the complexity added by the BOPS customers. This she expected. What she wasn’t expecting were the latest sales results; according to the most recent data (Table 1), since the introduction of BOPS sales had declined significantly in both the online and B&M channels.

<a><img src = "sales.JPG"></a>

She had two weeks to make sense of the data and provide concrete recommendations at the executive meeting. Despite the poor results, she felt instinctively that BOPS added value for Home and Kitchen customers. She was also uncomfortable abandoning BOPS so quickly. For one thing, it took a significant development effort to integrate the data of the B&M stores and the online store, and a lot of money, time, and effort had been invested to make this happen. She hated to see that investment go to waste. Also, many other major retailers had started to offer similar options in the last few years and Schwarz felt Home and Kitchen needed BOPS to remain competitive. But she knew she had some explaining to do given the poor sales results.

To better understand the perplexing results of the BOPS trial, Schwarz asked for a more detailed breakdown of sales data by week and by location. These weekly sales data cover a six-month period before the launch of BOPS (April 11–October 11, 2011) and a six-month period after the launch (October 11, 2011–April 11, 2012). The online sales are disaggregated by the geographical region from which the website customer originated. Toward this end, North America was divided into 210 Designated Market Areas (DMAs). A standard geographical unit in marketing, a DMA is a region of the country in which radio and television stations in the major city of the area are seen in homes and households as defined by Nielsen Media research. B&M sales are disaggregated by store.

With these more detailed data in hand, Schwarz wondered what new information it might
reveal and how she could use the data to gain better insight into the true effect of BOPS.
Ultimately, despite her conviction about the benefits of BOPS and her personal stake in the
project, she needed to determine objectively if BOPS was helping or hurting the company.

# **Data Visualization**

```python
import matplotlib.pyplot as plt
```

```python
import matplotlib
```

```python
from scipy.stats import pearsonr
```

## B&M Sales

Compare the total sales per week, before and after the BOPS deployment:

```python
bm_data_total_sales_before = bm_data_before.groupby(['week']).sum()['sales'].reset_index()
```

```python
bm_data_total_sales_after = bm_data_after.groupby(['week']).sum()['sales'].reset_index()
```

```python
plt.plot(bm_data_total_sales_before['sales'], label = "Total Sales Before")
plt.plot(bm_data_total_sales_after['sales'], label = "Total Sales After")
plt.xlabel('Week')
plt.ylabel('Total Sales per tens of millions')
plt.title('Sales by Week Before/After BOPS')
plt.rcParams["figure.figsize"] = (10,5)
plt.legend()
plt.show()
```

```python
plt.scatter(bm_data_total_sales_before['sales'], bm_data_total_sales_after['sales'])
plt.show()
```

```python
corr, _ = pearsonr(bm_data_total_sales_before['sales'], bm_data_total_sales_after['sales'])
print('Pearsons correlation: %.3f' % corr)
```

```python
plt.plot(bm_data_total_sales_after['sales']-bm_data_total_sales_before['sales'], label = "Diff")
plt.xlabel('Week')
plt.ylabel('Total Sales per tens of millions')
plt.title('Difference by Week Before/After BOPS')
plt.rcParams["figure.figsize"] = (10,5)
plt.legend()
plt.show()
```

Let's go further, separate the data for Stores IN **USA** and **Canada**:

```python
#For USA
bm_data_before_usa = bm_data_before[bm_data_before['usa']==1]
bm_data_after_usa = bm_data_after[bm_data_after['usa']==1]
bm_data_usa_sales_before = bm_data_before_usa.groupby(['week']).sum()['sales'].reset_index()
bm_data_usa_sales_after = bm_data_after_usa.groupby(['week']).sum()['sales'].reset_index()
#For Canada
bm_data_before_canada = bm_data_before[bm_data_before['usa']==0]
bm_data_after_canada = bm_data_after[bm_data_after['usa']==0]
bm_data_canada_sales_before = bm_data_before_canada.groupby(['week']).sum()['sales'].reset_index()
bm_data_canada_sales_after = bm_data_after_canada.groupby(['week']).sum()['sales'].reset_index()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
fig.suptitle('Total Sales Per Country')
ax1.plot(bm_data_usa_sales_before['sales'], label = "Total Sales Before")
ax1.plot(bm_data_usa_sales_after['sales'], label = "Total Sales After")
ax1.set_title('USA')
ax1.legend()

ax2.plot(bm_data_canada_sales_before['sales'], label = "Total Sales Before")
ax2.plot(bm_data_canada_sales_after['sales'], label = "Total Sales After")
ax2.set_title('Canada')
ax2.legend()

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
fig.suptitle('Total Diff. Sales Per Country')
ax1.plot(bm_data_usa_sales_after['sales']-bm_data_usa_sales_before['sales'], label = "Diff")
ax1.set_title('USA')
ax1.legend()
ax2.plot(bm_data_canada_sales_after['sales']-bm_data_canada_sales_before['sales'], label = "Diff")
ax2.set_title('Canada')
ax2.legend()

plt.show()
```

## Online Sales

Let's go further, separate the data for DMA **within 50 miles of a store** and **further than 50 miles**:

```python
#in a ratio of 50 miles
online_data_before_close = online_data_before[online_data_before['close']==1]
online_data_after_close = online_data_after[online_data_after['close']==1]
online_data_close_sales_before = online_data_before_close.groupby(['week']).sum()['sales'].reset_index()
online_data_close_sales_after = online_data_after_close.groupby(['week']).sum()['sales'].reset_index()
#more than 50 miles
online_data_before_far = online_data_before[online_data_before['close']==0]
online_data_after_far = online_data_after[online_data_after['close']==0]
online_data_far_sales_before = online_data_before_far.groupby(['week']).sum()['sales'].reset_index()
online_data_far_sales_after = online_data_after_far.groupby(['week']).sum()['sales'].reset_index()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
fig.suptitle('Total Sales Per Distance')
ax1.plot(online_data_close_sales_before['sales'], label = "Total Sales Before")
ax1.plot(online_data_close_sales_after['sales'], label = "Total Sales After")
ax1.set_title('Within 50 miles of a Store')
ax1.legend()

ax2.plot(online_data_far_sales_before['sales'], label = "Total Sales Before")
ax2.plot(online_data_far_sales_after['sales'], label = "Total Sales After")
ax2.set_title('Further than 50 miles')
ax2.legend()

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
fig.suptitle('Difference on Total Sales Per Distance')
ax1.plot(online_data_close_sales_after['sales']-online_data_close_sales_before['sales'], label = "Diff")
ax1.set_title('Within 50 miles of a Store')
ax1.legend()

ax2.plot(online_data_far_sales_after['sales']-online_data_far_sales_before['sales'], label = "Total Sales Before")
ax2.set_title('Further than 50 miles')
ax2.legend()

plt.show()
```
